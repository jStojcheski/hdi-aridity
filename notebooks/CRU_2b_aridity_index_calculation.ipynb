### CRU: Aridity Index Calculation

```python
import pandas as pd
```

#### Set input/output folder

```python
data_folder  = '../data/CRU_2a_iso_standardization'

periods = ['annually', 'monthly', 'quarterly']
```

#### Calculate aridity index

$\text{Aridity Index} = \frac{\text{Precipitation}}{\text{Potential Evapotranspiration}}$

#### Read PRE (precipitation) and PET (potential evapotranspiration) values; Output to csv format

```python
for period in periods:
    pre_df = pd.read_csv(f'{data_folder}/{period}/Precipitation (PRE) [mm_month].csv',
                         index_col='Country')
    pet_df = pd.read_csv(f'{data_folder}/{period}/Potential Evapotranspiration (PET) [mm_month].csv',
                         index_col='Country')
    
    ai_df = pre_df / pet_df
    display(ai_df.head(10))
    
    ai_df.to_csv(f'{data_folder}/{period}/Aridity Index (AI).csv')
```
