### HDR: Country names ISO standardization

```python
import os
import pandas as pd
```

#### Set input/output folder

```python
data_in  = '../data/HDR_1a_delete_annotations'
data_out = '../data/HDR_1b_iso_standardization'

if not os.path.exists(data_out):
    os.mkdir(data_out)
```

#### Read table of country names and ISO codes

```python
iso_df = pd.read_csv('../data/ISO_country_codes.csv',
                     index_col='HDR Country Name')
iso_df = iso_df.drop('CRU Country Name', axis=1)
iso_df[::10]
```

#### Read data, change country names to ISO codes and output data to csv format

```python
all_attrs = dict()

for dim in os.listdir(data_in):
    dim_path = f'{data_in}/{dim}'
    if os.path.isdir(dim_path):
        
        if dim not in os.listdir(data_out):
            os.mkdir(f'{data_out}/{dim}')
            
        for attr_file in os.listdir(dim_path):
            if not os.path.isdir(attr_file):
                attr = f'[{dim}] {attr_file[:-4]}'
                all_attrs[attr] = pd.read_csv(f'{dim_path}/{attr_file}', index_col='Country')
                all_attrs[attr].index = iso_df['alpha-3'][all_attrs[attr].index]
                all_attrs[attr].to_csv(f'{data_out}/{dim}/{attr_file}', index_label='Country')
                
    else:
        attr = dim[:-4]
        all_attrs[attr] = pd.read_csv(f'{dim_path}', index_col='Country')
        all_attrs[attr].index = iso_df['alpha-3'][all_attrs[attr].index]
        all_attrs[attr].to_csv(f'{data_out}/{attr}.csv', index_label='Country')
```
