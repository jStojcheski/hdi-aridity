### HDR: Weighted Stochastic Block Model (WSBM) analysis

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import scipy as sp
import seaborn

%matplotlib inline
```

#### Settings

```python
year_s = 2010
year_f = 2016
years = list(np.arange(year_s, year_f+1))

period = f'{year_s}_{year_f}'

mean_type = 'mean'  # ['mean', 'median', 'wmean']
```

#### Set input/output folder

```python
settings = 'full/exp_0'

if settings.split('/')[0] == 'full':
    data_gf_in = f'../data/HDR_4a_graph_formation/{mean_type}'
else:
    data_gf_in = f'../data/HDR_4b_sparsity_threshold/{mean_type}'

data_wsbm_in = f'../data/HDR_4c_wsbm/01_merged/{settings}'
data_mt      =  '../data/HDR_mutual_attributes_and_countries'

data_ai_in  =  '../data/CRU_3_aggregate_attr_by_period/annually'
data_hd_in  =  '../data/HDR_1c_aggregate_attr_by_year'

data_out = f'../data/HDR_4c_wsbm/02_results/{settings}'
if not os.path.exists(data_out):
    os.makedirs(data_out)

plot_out = f'../plots/HDR_4c_wsbm/{settings}'
if not os.path.exists(plot_out):
    os.makedirs(plot_out)
```

#### Read input data

```python
gf = dict()

wsbm_labels = dict()
wsbm_logevd = pd.DataFrame()

ai = dict()
hd = dict()

for year in years:  # years
    # read graphs (both fully-connected and sparse)
    gf[year] = pd.read_csv(f'{data_gf_in}/l1_{year}.csv', index_col='Country')
    
    # read wsbm labels and log-evidence values
    wsbm_labels[year] = pd.read_csv(f'{data_wsbm_in}/{year}_labels.csv', index_col='Country')
    wsbm_logevd[year] = pd.read_csv(f'{data_wsbm_in}/{year}_logevd.csv', index_col='Measure').transpose()['Log-Likelihood']
    
    # read Aridity Index (AI) and Human Development Index (HDI) data
    ai[year] = pd.read_csv(f'{data_ai_in}/{year}.csv', index_col='Country')['Aridity Index (AI)']
    hd[year] = pd.read_csv(f'{data_hd_in}/{year}.csv', index_col='Country')['Human Development Index (HDI)']
    
    print(f'{year}:', (wsbm_labels[year].columns.astype('int') < wsbm_labels[year].max(axis=0)).sum())
```

#### Edge-weight distribution

```python
def plot_distr(graphs, years, n=10000, title='hist'):
    f, axs = plt.subplots(1, len(years), figsize=(16, 3), sharey='row')  # sharex='col', 
    f.subplots_adjust(hspace=0, wspace=0)

    for i in range(len(years)):
        year = years[i]
        hist_data = graphs[year].replace(0, np.nan).values.flatten()
    #     print(len(hist_data))

        hist_data = hist_data[~np.isnan(hist_data)]
    #     print(len(hist_data))

        g1, g2, g3 = sp.stats.gamma.fit(hist_data)
#         e1, e2     = sp.stats.expon.fit(hist_data)  # assumes lambda = 1

        x = np.arange(0, 1 + 1/n, 1/n)
        
        ks_results = sp.stats.kstest(hist_data, "gamma", (g1, g2, g3))

        print(f'{year} | GAMMA: ({g1:.3f} {g2:.3f} {g3:.3f}) | p={ks_results.pvalue}')
#         print(f'EXPON: {e1} {e2}')
#         print(f'       {sp.stats.kstest(hist_data, "expon", (e1, e2))}')

#         plt.figure(figsize=(16, 9))
        axs[i].hist(hist_data, bins=10, density=True)
        axs[i].plot(x, sp.stats.gamma.pdf(x, g1, g2, g3))
    #     plt.plot(x, sp.stats.expon.pdf(x, e1, e2))
        axs[i].set_title(f'{year}')
        axs[i].set_xlim(0, 1)
        axs[i].set_xticks([])
#         axs[i].set_xticklabels(np.arange(0, 1.01, 0.25), rotation=45)
    plt.savefig(f'../plots/HDR_4c_wsbm/{title}')
    plt.show()
```

##### Complete graph

```python
plot_distr(gf, years, title='full_gamma')
```

##### Clustering

```python
wsbm_max = pd.DataFrame()
ks = pd.Series(index=years, dtype=int)

for year, idx in max_idx.items():
    wsbm_max[year] = wsbm_labels[year][idx]
    ks[year] = len(set(wsbm_max[year]))
    print(f'{year}: {ks[year]}')

wsbm_max.transpose()
```

#### Plotting results

```python
f, axs_acc = plt.subplots(len(years), 3, figsize=(16, 25))  # sharex='col', sharey='col'

for i in range(len(years)):
    year = years[i]
    
    clusters = wsbm_max[year].sort_values()
    
    clusters_df = pd.concat([clusters, ai[year], hd[year]], axis=1, sort=True).dropna()
    clusters_df.columns = ['Cluster', 'AI', 'HDI']
    clusters_df = clusters_df.sort_values(by=['Cluster', 'HDI'])

    # Heatmaps
    seaborn.heatmap((1 - gf[year]).reindex(index=clusters.index,
                                           columns=clusters.index),
                    ax=axs_acc[i][0], cbar=False, cmap='Oranges', vmax=1, vmin=0)
    axs_acc[i][0].set_title(f'{year} (K = {ks[year]})')
    
    axs_acc[i][0].set_xticks([])
    axs_acc[i][0].set_yticks([])
    axs_acc[i][0].set_xlabel('')
    axs_acc[i][0].set_ylabel('')

    offset = 0
    for j in range(ks[year]):
        block_size = len(clusters[clusters == j+1])
        idx =  block_size + offset
        
        axs_acc[i][0].hlines(idx, xmin=offset, xmax=idx)
        axs_acc[i][0].vlines(idx, ymin=offset, ymax=idx)
        
        axs_acc[i][0].hlines(offset, xmin=offset, xmax=idx)
        axs_acc[i][0].vlines(offset, ymin=offset, ymax=idx)
        
        offset = idx

    plt.figure(figsize=(8, 6))
    seaborn.heatmap((1 - gf[year]).reindex(index=clusters.index,
                                           columns=clusters.index),
                    cbar=False, cmap='Oranges', vmax=1, vmin=0)
    
    offset = 0
    for j in range(ks[year]):
        block_size = len(clusters[clusters == j+1])
        idx =  block_size + offset
        
        plt.hlines(idx, xmin=offset, xmax=idx)
        plt.vlines(idx, ymin=offset, ymax=idx)
        
        plt.hlines(offset, xmin=offset, xmax=idx)
        plt.vlines(offset, ymin=offset, ymax=idx)
        
        offset = idx
        
    plt.savefig(f'{plot_out}/heatmap_{year}')
    plt.close()
    
    # Human Development Index groups
    axs_acc[i][1].plot(clusters_df['Cluster'], clusters_df['HDI'], 'o', alpha=0.4)
    axs_acc[i][1].set_title(f'HDI {year}')
    axs_acc[i][1].set_xticks([])
    
    plt.figure(figsize=(16, 9))
    plt.plot(clusters_df['Cluster'], clusters_df['HDI'], 'o', alpha=0.4)
    plt.title(f'HDI {year}')
    plt.savefig(f'{plot_out}/hdi_{year}')
    plt.close()

    # Aridity Index groups
    axs_acc[i][2].plot(clusters_df['Cluster'], clusters_df['AI'], 'o', alpha=0.4)
    axs_acc[i][2].set_title(f'AI {year}')
    axs_acc[i][2].set_xticks([])

    plt.figure(figsize=(16, 9))
    plt.plot(clusters_df['Cluster'], clusters_df['AI'], 'o', alpha=0.4)
    plt.title(f'AI {year}')
    plt.savefig(f'{plot_out}/ai_{year}')
    plt.close()

f.savefig(f'{plot_out}/wsbm_results')
f.show()
```

#### Modularity

```python
# %load '../snippets/modularity.py'
def modularity(df, groups):
    labels = set(groups)

    edge_set = set()
    value = 0

    for i in labels:
        for j in labels:
            cs1 = groups[groups == i].index
            cs2 = groups[groups == j].index

            for c1 in cs1:
                for c2 in cs2:
                    edge = f'{c1}_{c2}' if c1 < c2 else f'{c2}_{c1}'

                    if edge not in edge_set:
                        if i == j:
                            value += df.loc[c1, c2]
                        else:
                            value -= df.loc[c1, c2]

                        edge_set.add(edge)

    return value
```

```python
wsbm_df = pd.DataFrame([modularity(gf[year], wsbm_max[year]) for year in years],
                       columns=['WSBM'], index=years)
display(wsbm_df)

wsbm_df.to_csv(f'{data_out}/wsbm.csv', index_label='Year')
```
