```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd

from pprint import pprint

%matplotlib inline
```

#### Read input data

```python
mod_df = pd.read_csv('../data/HDR_4c_mean_full.csv', index_col='Year')
mod_df = mod_df.transpose()
mod_df.columns.name = None

idx = list('AP (' + mod_df.index[:-1] + ')')
idx.append(mod_df.index[-1] + ' (Best)')
mod_df.index = idx
pprint(list(mod_df.index))

years = mod_df.columns

mod_df = mod_df.transpose()
display(mod_df.transpose())
```

#### Get best models for each year

```python
min_idx = mod_df.idxmin(axis=1)
min_idx
```

#### Plotting
